```python
import numpy as np
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import time
```

```python
def show_images(image, num_row=2, num_col=5):
    image_size = int(np.sqrt(image.shape[-1]))
    image = np.reshape(image, (image.shape[0], image_size, image_size))
    fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col, 2*num_row))
    for i in range(num_row*num_col):
        ax = axes[i // num_col, i % num_col]
        ax.imshow(image[i], cmap='gray', vmin=0, vmax=1)
        ax.axis('off')
    plt.tight_layout()
    plt.show()

def one_hot(x, k, dtype=np.float32):
    return np.array(x[:, None] == np.arange(k), dtype)
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()


x_train= x_train / 255.0
x_test = x_test / 255.0

num_labels = 10
y_train=one_hot(y_train.astype('int32'),num_labels)
y_test=one_hot(y_test.astype('int32'),num_labels)


x_train = x_train.reshape(60000,28*28)
x_test = x_test.reshape(10000,28*28)

print("Training data: {} {}".format(x_train.shape, y_train.shape))
print("Test data: {} {}".format(x_test.shape, y_test.shape))
show_images(x_train)
```

```python
class DeepNeuralNetwork:
    def __init__(self, sizes, activation='sigmoid', reg_type=None, reg_lambda=0.01):
        self.sizes = sizes

        if activation == 'relu':
            self.activation = self.relu
        elif activation == 'sigmoid':
            self.activation = self.sigmoid
        else:
            raise ValueError("Activation function is currently not support, please use 'relu' or 'sigmoid' instead.")

        self.reg_type = reg_type
        self.reg_lambda = reg_lambda

        self.params = self.initialize()
        self.cache = {}

    def relu(self, x, derivative=False):
        '''
        Derivative of ReLU is a bit more complicated since it is not differentiable at x = 0

        Forward path:
        relu(x) = max(0, x)
        In other word,
        relu(x) = 0, if x < 0
                = x, if x >= 0

        Backward path:
        ∇relu(x) = 0, if x < 0
                  = 1, if x >=0
        '''

        if derivative:
            return (x > 0).astype(float)
        return np.maximum(0, x)


    def sigmoid(self, x, derivative=False):

        '''
            Forward path:
            σ(x) = 1 / 1+exp(-z)

            Backward path:
            ∇σ(x) = exp(-z) / (1+exp(-z))^2
        '''

        sig = 1 / (1 + np.exp(-x))
        if derivative:
            return sig * (1 - sig)
        return sig

    def softmax(self, x):

        '''
            softmax(x) = exp(x) / ∑exp(x)
        '''

        exps = np.exp(x - np.max(x, axis=0, keepdims=True))
        return exps / np.sum(exps, axis=0, keepdims=True)

    def initialize(self):
        # number of nodes in each layer
        input_layer=self.sizes[0]
        hidden_layer=self.sizes[1]
        output_layer=self.sizes[2]

        params = {
            "W1": np.random.randn(hidden_layer, input_layer) * np.sqrt(1./input_layer),
            "b1": np.zeros((hidden_layer, 1)) * np.sqrt(1./input_layer),
            "W2": np.random.randn(output_layer, hidden_layer) * np.sqrt(1./hidden_layer),
            "b2": np.zeros((output_layer, 1)) * np.sqrt(1./hidden_layer)
        }
        return params


    def initialize_optimizer(self, optimizer='sgd'):
        if optimizer in ['momentum', 'adam', 'rmsprop']:
            self.velocity = {key: np.zeros_like(value) for key, value in self.params.items()}
        if optimizer == 'adam':
            self.moment = {key: np.zeros_like(value) for key, value in self.params.items()}
            self.t = 0

    def feed_forward(self, x):
        self.cache["X"] = x
        self.cache["Z1"] = np.dot(self.params["W1"], x.T) + self.params["b1"]
        self.cache["A1"] = self.activation(self.cache["Z1"])
        self.cache["Z2"] = np.dot(self.params["W2"], self.cache["A1"]) + self.params["b2"]
        self.cache["A2"] = self.softmax(self.cache["Z2"])
        return self.cache["A2"]

    def back_propagate(self, y, output):
        '''
            This is the backpropagation algorithm, for calculating the updates
            of the neural network's parameters.

            Note: There is a stability issue that causes warnings. This is
                  caused  by the dot and multiply operations on the huge arrays.

                  RuntimeWarning: invalid value encountered in true_divide
                  RuntimeWarning: overflow encountered in exp
                  RuntimeWarning: overflow encountered in square
        '''
        batch_size = y.shape[0]
        dZ2 = output - y.T
        dW2 = (1. / batch_size) * np.dot(dZ2, self.cache["A1"].T)
        db2 = (1. / batch_size) * np.sum(dZ2, axis=1, keepdims=True)
        dA1 = np.dot(self.params["W2"].T, dZ2)
        dZ1 = dA1 * self.activation(self.cache["Z1"], derivative=True)
        dW1 = (1. / batch_size) * np.dot(dZ1, self.cache["X"])
        db1 = (1. / batch_size) * np.sum(dZ1, axis=1, keepdims=True)

        if self.reg_type == 'L2':
            dW1 += (self.reg_lambda / batch_size) * self.params["W1"]
            dW2 += (self.reg_lambda / batch_size) * self.params["W2"]
        elif self.reg_type == 'L1':
            dW1 += (self.reg_lambda / batch_size) * np.sign(self.params["W1"])
            dW2 += (self.reg_lambda / batch_size) * np.sign(self.params["W2"])

        return {"W1": dW1, "b1": db1, "W2": dW2, "b2": db2}

    def cross_entropy_loss(self, y, output):
        output = np.clip(output, 1e-10, 1 - 1e-10)
        loss = -np.mean(np.sum(y * np.log(output.T), axis=1))

        if self.reg_type == 'L2':
            loss += (self.reg_lambda / (2 * y.shape[0])) * (np.sum(np.square(self.params["W1"])) + np.sum(np.square(self.params["W2"])))
        elif self.reg_type == 'L1':
            loss += (self.reg_lambda / y.shape[0]) * (np.sum(np.abs(self.params["W1"])) + np.sum(np.abs(self.params["W2"])))

        return loss

    def optimize(self, optimizer, l_rate=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
        if optimizer == 'sgd':
            for key in self.params:
                self.params[key] -= l_rate * self.grads[key]
        elif optimizer == 'momentum':
            for key in self.params:
                self.velocity[key] = beta1 * self.velocity[key] + (1 - beta1) * self.grads[key]
                self.params[key] -= l_rate * self.velocity[key]
        elif optimizer == 'rmsprop':
            for key in self.params:
                self.velocity[key] = beta1 * self.velocity[key] + (1 - beta1) * (self.grads[key] ** 2)
                self.params[key] -= l_rate * self.grads[key] / (np.sqrt(self.velocity[key]) + epsilon)
        elif optimizer == 'adam':
            self.t += 1
            for key in self.params:
                self.moment[key] = beta1 * self.moment[key] + (1 - beta1) * self.grads[key]
                self.velocity[key] = beta2 * self.velocity[key] + (1 - beta2) * (self.grads[key] ** 2)
                m_hat = self.moment[key] / (1 - beta1 ** self.t)
                v_hat = self.velocity[key] / (1 - beta2 ** self.t)
                self.params[key] -= l_rate * m_hat / (np.sqrt(v_hat) + epsilon)

    def accuracy(self, y, output):
        return np.mean(np.argmax(y, axis=-1) == np.argmax(output.T, axis=-1))

    def train(self, x_train, y_train, x_test, y_test, epochs=10, batch_size=64, optimizer='momentum', l_rate=0.01, beta=0.9):
        self.initialize_optimizer(optimizer)
        for epoch in range(epochs):
            permutation = np.random.permutation(x_train.shape[0])
            x_train_shuffled = x_train[permutation]
            y_train_shuffled = y_train[permutation]
            for batch in range(0, x_train.shape[0], batch_size):
                x = x_train_shuffled[batch:batch + batch_size]
                y = y_train_shuffled[batch:batch + batch_size]
                output = self.feed_forward(x)
                self.grads = self.back_propagate(y, output)
                self.optimize(optimizer, l_rate, beta)
            train_output = self.feed_forward(x_train)
            test_output = self.feed_forward(x_test)
            train_acc = self.accuracy(y_train, train_output)
            test_acc = self.accuracy(y_test, test_output)
            train_loss = self.cross_entropy_loss(y_train, train_output)
            test_loss = self.cross_entropy_loss(y_test, test_output)
            print(f"Epoch {epoch + 1}, Egitim Accuracy: {train_acc:.2f}, Test Accuracy: {test_acc:.2f}, Egitim Loss: {train_loss:.2f}, Test Loss: {test_loss:.2f}")

```

```python

epochs = 10
batch_size = 64
optimizers = ['adam', 'rmsprop', 'sgd', 'momentum']
results = {}

for opt in optimizers:
    print(f"Sonuçlar (L2 Regularizasyon, {opt} optimizer):\n")
    model = DeepNeuralNetwork(sizes=[784, 64, 10], activation='relu', reg_type='L2', reg_lambda=0.01)
    model.initialize_optimizer(opt)
    start_time = time.time()
    model.train(x_train, y_train, x_test, y_test, optimizer=opt, l_rate=0.01, beta=0.9)
    elapsed_time = time.time() - start_time
    results[f"{opt}_L2"] = {
        "egitim_accuracy": model.accuracy(y_train, model.feed_forward(x_train)),
        "test_accuracy": model.accuracy(y_test, model.feed_forward(x_test)),
        "zaman": elapsed_time
    }
```

```python


class DeepNeuralNetwork:
    def __init__(self, sizes, activation='sigmoid'):
        self.sizes = sizes

        if activation == 'relu':
            self.activation = self.relu
        elif activation == 'sigmoid':
            self.activation = self.sigmoid
        else:
            raise ValueError("Activation function is currently not supported, please use 'relu' or 'sigmoid' instead.")

        self.params = self.initialize()
        self.cache = {}

    def relu(self, x, derivative=False):
        '''
        Derivative of ReLU is a bit more complicated since it is not differentiable at x = 0

        Forward path:
        relu(x) = max(0, x)
        In other word,
        relu(x) = 0, if x < 0
                = x, if x >= 0

        Backward path:
        ∇relu(x) = 0, if x < 0
                  = 1, if x >=0
        '''
        if derivative:
            return (x > 0).astype(float)
        return np.maximum(0, x)

    def sigmoid(self, x, derivative=False):

        '''
            Forward path:
            σ(x) = 1 / 1+exp(-z)

            Backward path:
            ∇σ(x) = exp(-z) / (1+exp(-z))^2
        '''
        sig = 1 / (1 + np.exp(-x))
        if derivative:
            return sig * (1 - sig)
        return sig

    def softmax(self, x):

        '''
            softmax(x) = exp(x) / ∑exp(x)
        '''
        exps = np.exp(x - np.max(x, axis=0, keepdims=True))
        return exps / np.sum(exps, axis=0, keepdims=True)

    def initialize(self):
        input_layer, hidden_layer, output_layer = self.sizes
        return {
            "W1": np.random.randn(hidden_layer, input_layer) * np.sqrt(1. / input_layer),
            "b1": np.zeros((hidden_layer, 1)),
            "W2": np.random.randn(output_layer, hidden_layer) * np.sqrt(1. / hidden_layer),
            "b2": np.zeros((output_layer, 1))
        }

    def initialize_optimizer(self, optimizer='sgd'):
        if optimizer in ['momentum', 'adam', 'rmsprop']:
            self.velocity = {key: np.zeros_like(value) for key, value in self.params.items()}
        if optimizer == 'adam':
            self.moment = {key: np.zeros_like(value) for key, value in self.params.items()}
            self.t = 0

    def feed_forward(self, x):
        self.cache["X"] = x
        self.cache["Z1"] = np.dot(self.params["W1"], x.T) + self.params["b1"]
        self.cache["A1"] = self.activation(self.cache["Z1"])
        self.cache["Z2"] = np.dot(self.params["W2"], self.cache["A1"]) + self.params["b2"]
        self.cache["A2"] = self.softmax(self.cache["Z2"])
        return self.cache["A2"]

    def back_propagate(self, y, output):
        '''
            This is the backpropagation algorithm, for calculating the updates
            of the neural network's parameters.

            Note: There is a stability issue that causes warnings. This is
                  caused  by the dot and multiply operations on the huge arrays.

                  RuntimeWarning: invalid value encountered in true_divide
                  RuntimeWarning: overflow encountered in exp
                  RuntimeWarning: overflow encountered in square
        '''
        batch_size = y.shape[0]
        dZ2 = output - y.T
        dW2 = (1. / batch_size) * np.dot(dZ2, self.cache["A1"].T)
        db2 = (1. / batch_size) * np.sum(dZ2, axis=1, keepdims=True)
        dA1 = np.dot(self.params["W2"].T, dZ2)
        dZ1 = dA1 * self.activation(self.cache["Z1"], derivative=True)
        dW1 = (1. / batch_size) * np.dot(dZ1, self.cache["X"])
        db1 = (1. / batch_size) * np.sum(dZ1, axis=1, keepdims=True)

        return {"W1": dW1, "b1": db1, "W2": dW2, "b2": db2}

    def cross_entropy_loss(self, y, output):
        output = np.clip(output, 1e-10, 1 - 1e-10)
        loss = -np.mean(np.sum(y * np.log(output.T), axis=1))
        return loss

    def optimize(self, optimizer, l_rate=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
        if optimizer == 'sgd':
            for key in self.params:
                self.params[key] -= l_rate * self.grads[key]
        elif optimizer == 'momentum':
            for key in self.params:
                self.velocity[key] = beta1 * self.velocity[key] + (1 - beta1) * self.grads[key]
                self.params[key] -= l_rate * self.velocity[key]
        elif optimizer == 'rmsprop':
            for key in self.params:
                self.velocity[key] = beta1 * self.velocity[key] + (1 - beta1) * (self.grads[key] ** 2)
                self.params[key] -= l_rate * self.grads[key] / (np.sqrt(self.velocity[key]) + epsilon)
        elif optimizer == 'adam':
            self.t += 1
            for key in self.params:
                self.moment[key] = beta1 * self.moment[key] + (1 - beta1) * self.grads[key]
                self.velocity[key] = beta2 * self.velocity[key] + (1 - beta2) * (self.grads[key] ** 2)
                m_hat = self.moment[key] / (1 - beta1 ** self.t)
                v_hat = self.velocity[key] / (1 - beta2 ** self.t)
                self.params[key] -= l_rate * m_hat / (np.sqrt(v_hat) + epsilon)

    def accuracy(self, y, output):
        return np.mean(np.argmax(y, axis=-1) == np.argmax(output.T, axis=-1))

    def train(self, x_train, y_train, x_test, y_test, epochs=10, batch_size=64, optimizer='momentum', l_rate=0.01, beta=0.9):
        self.initialize_optimizer(optimizer)
        for epoch in range(epochs):
            permutation = np.random.permutation(x_train.shape[0])
            x_train_shuffled = x_train[permutation]
            y_train_shuffled = y_train[permutation]
            for batch in range(0, x_train.shape[0], batch_size):
                x = x_train_shuffled[batch:batch + batch_size]
                y = y_train_shuffled[batch:batch + batch_size]
                output = self.feed_forward(x)
                self.grads = self.back_propagate(y, output)
                self.optimize(optimizer, l_rate, beta)
            train_output = self.feed_forward(x_train)
            test_output = self.feed_forward(x_test)
            train_acc = self.accuracy(y_train, train_output)
            test_acc = self.accuracy(y_test, test_output)
            train_loss = self.cross_entropy_loss(y_train, train_output)
            test_loss = self.cross_entropy_loss(y_test, test_output)
            print(f"Epoch {epoch + 1}, egitim Accuracy: {train_acc:.2f}, Test Accuracy: {test_acc:.2f}, egitim Loss: {train_loss:.2f}, Test Loss: {test_loss:.2f}")
```
